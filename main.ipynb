```python
from tqdm import tqdm
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')
# pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('./data/final_data.csv')
df.head(3)
```

```python
dfg = df.loc[df['country']!='Global'].groupby(['sales_year', 'country'])['yearly'].sum().reset_index()
dfg['yearly_mln'] = dfg['yearly'] / 1000000

plt.figure(figsize=(7, 5))
plt.ylabel('Миллионов штук')
plt.xlabel('Год')
plt.title('Продажи видеоигр в рознице по количеству проданных штук')
sns.lineplot(x=dfg['sales_year'], y=dfg['yearly_mln'], data=dfg, hue='country')
# plt.legend(loc='upper right')
plt.show()
```
